```python
batch_size = 32
time_steps = 48

z_dims_t = 10
y_dims = z_dims_t
Dx = 20

dname = "SineImage"
test = "cot"

g_filter_size = 32
d_filter_size = g_filter_size
g_state_size = 32
d_state_size = 32
p_filter_size = 0

output_activation = 'sigmoid'
z_dims = z_dims_t * time_steps

sinkhorn_eps = 10
sinkhorn_l   = 100
reg_penalty = 10.0
nlstm = 1
seed = 42
lr = 1e-3
dist_z = tfp.distributions.Uniform(-1, 1)
dist_y = tfp.distributions.Uniform(-1, 1)

def sample_generator(gen, n, batch_size=batch_size, z_dims=z_dims):
    nbatch = n // batch_size + 1
    data = []
    for ni in range(nbatch):
        z = dist_z.sample([batch_size, time_steps, z_dims_t])
        y = dist_y.sample([batch_size, y_dims])
        data += gen.call(z, y, training=False).numpy(),
    data = np.concatenate(data, 0)
    return data[:n]
```

```python
plt.style.use("default")
tests = [ "W4", "WaveGAN"]
def plot_corr(axes, data):
    sample = np.asarray(data)
    autocs = []
    for i in range(64):
        autoc = np.corrcoef(sample[...,i].T + np.random.randn(*(sample[...,i].T).shape)*0.000001)[:,0]
        autocs += autoc,
        axes[0].plot(range(1,101), autoc, color=f"C{1}", ls="-", lw=2)
    axes[0].set_ylim([-0.3,1.0])
    
    s_corr = np.corrcoef(sample.reshape(-1,sample.shape[-1]).T)
    axes[1].imshow(s_corr, vmin=-0.7,vmax=1)
    
    f = np.abs(np.fft.fft(sample, axis=1))
    f = np.fft.fftshift(f,1)
    f = f.mean(0)
    axes[2].plot(np.arange(1,51),f[50:])
    axes[2].set_yscale("log")
#   axes[2].set_xscale("log")
    
    return np.asarray(autocs), s_corr, f


proper_names = ["real", "COT-GAN", "WaveGAN"]


fig, axes = plt.subplots(4,len(data_dict),figsize=(12,6))
ax = axes[:,0]
ax[0].plot(data_dict["real"][0])
ax[0].set_title("real")
autoc_real, sc_real, f_real = plot_corr(ax[1:], data_dist.test_data)
ax[1].set_title("0.0")
ax[2].set_title("0.0")
ax[3].set_title("0.0")

for ni, name in enumerate(tests):
    ax = axes[:,ni+1]
    ax[0].plot(data_dict[name][0])
    ax[0].set_title(proper_names[ni+1])
    autoc, sc, f = plot_corr(ax[1:], data_dict[name])
    autoc_diff = np.sum(np.abs(autoc-autoc_real))
    sc_diff = np.sum(np.abs(sc-sc_real))
    f_diff  = np.sum(np.abs(f-f_real))
    ax[1].set_title(f"{autoc_diff:.1f}")
    ax[2].set_title(f"{sc_diff:.1f}")
    ax[3].set_title(f"{f_diff:.1f}")

fig.tight_layout(pad=0.1)
```
